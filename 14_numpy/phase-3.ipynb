```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id,2021, 2022, 2023, 2024]

sales_data=np.array([
  [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
  [2, 120000, 140000, 160000, 190000],  # Beijing Bites
  [3, 200000, 230000, 260000, 300000],  # Pizza Hub
  [4, 180000, 210000, 240000, 270000],  # Burger Point
  [5, 160000, 185000, 205000, 230000]   # Chai Point
  ])  

print("=== Zomato sales analysis ===")
print("\nSales data shape: ",sales_data.shape)
print("\n Data of first 3 restaurants:\n",sales_data[:3,])
```

```python
# Total sales per year

print(np.sum(sales_data,axis=0))
print(np.sum(sales_data,axis=1))

print(np.sum(sales_data[:,1:],axis=0))
print(np.sum(sales_data[:,1:],axis=1))

# Minimum sales per restaurant
min_sales = np.min(sales_data[:,1:],axis=1)
print(min_sales)

# Maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)

# Average sales per restaurant
avg_sales=np.mean(sales_data[:,1:],axis=1)
print(avg_sales)

# cummulative sum
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cummulative sales across all restaurants ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

## Vector operations

```python
v1=np.array([1,2,3,4,5])
v2=np.array([6,7,8,9,10])

print("Vector addition: ",v1+v2)
print("\nVector Multiplication: ",v1*v2)
print("\nDot Product: ",np.dot(v1,v2))

angle=np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
print(angle)
```

## vectorize operation

```python
restaurant_types=np.array(['biryani','chinese','pizza','burger','cafe'])
vectorized_upper=np.vectorize(str.upper)
print("Vectorized upper: ",vectorized_upper(restaurant_types))
```
